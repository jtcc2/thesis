Here we give a visual comparison of the results of using the spectral ordering vs neighbour ordering vs greedy neighbour ordering, in our clustering algorithm.

```python
load("utils.sage")

p = 5569
L = [2, 3]
G = get_L_graph(p, L)

# First we plot the adjacency matrix after spectral ordering of vertices
S = get_spectral_ordering(G, use_numpy_over_sage=True)
M = G.adjacency_matrix(vertices = S)
M.plot()
```

```python
# pick random starting vertex for neighbour orderings
start_vertex = G.vertices()[randrange(0, len(G.vertices()))]
print("Using start vertex:  " + str(start_vertex))

# Plot for neighbour ordering
S2 = get_neighbour_ordering(G, [start_vertex])
M2 = G.adjacency_matrix(vertices = S2)
M2.plot()
```

```python
S3 = get_greedy_ordering(G, [start_vertex])
M3 = G.adjacency_matrix(vertices = S3)
M3.plot()
```

We plot these again, randomizing the ordering of vertices in the cut and its complement

```python
# spectral ordering
from random import shuffle
_, cut = fiedler(G, ordering=S)
cut_complement = vertex_complement(G, cut)
shuffle(cut)
shuffle(cut_complement)
M4 = G.adjacency_matrix(vertices = cut + cut_complement)
M4.plot()
```

```python
# neighbour ordering
_, cut = fiedler(G, ordering=S2)
cut_complement = vertex_complement(G, cut)
shuffle(cut)
shuffle(cut_complement)
M5 = G.adjacency_matrix(vertices = cut + cut_complement)
M5.plot()
```

```python
# greedy neighbour ordering
_, cut = fiedler(G, ordering=S3)
cut_complement = vertex_complement(G, cut)
shuffle(cut)
shuffle(cut_complement)
M6 = G.adjacency_matrix(vertices = cut + cut_complement)
M6.plot()
```
